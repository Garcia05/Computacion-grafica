```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img= cv2.imread('question_3.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_out= img.copy()


plt.axis("on")
plt.hist(img.ravel(),256,[0,256])
plt.show()


row, col = img.shape

f1 = 120#50
f2 = 250#100

for y in range(0,col):
    for x in range(0,row):
        if (f1 <= img[x,y] and img[x,y] <= f2):
            img_out[x,y] = 255

        else:
            img_out[x,y] = 0

cv2.imshow('Coverted Image',img_out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

inputt = cv2.imread('thresh3.png')

img2 = inputt.copy()
row,cols,channels = img2.shape

for i in range(row):
    for j in range (cols):
        if  133 < img2[i][j][0] and img2[i][j][0] < 186 and 158 < img2[i][j][1] and img2[i][j][1] < 201 and 211 < img2[i][j][2] and img2[i][j][2] < 238 or (img2[i][j][0] == [199 ,203, 233]).all():
            img2[i][j] = img2[i][j] 
        else:
            img2[i][j] = 0
            
edges = cv2.Canny(inputt,1,200)

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([inputt],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()

cv2.imwrite('output3.png',img2)

cv2.imshow('edges',edges)
cv2.imshow('img2',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
