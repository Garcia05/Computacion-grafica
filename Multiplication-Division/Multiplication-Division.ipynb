```python
# 4to de la practica blending
import cv2
import math
import cmath
import numpy as np
from matplotlib import pyplot as plt

def Blending(x,f1,f2):
    valor = x * f1 + (1-x)*f2
    if(valor>255):
        return 255
    if(valor<0):
        return 0
    return valor

    
img1= cv2.imread('add_10.jpg')
img2= cv2.imread('add_11.jpg')

row, cols ,channels = img1.shape
img = cv2.resize(img2,(int(cols),int(row)))
x = 1
for i in range(0,row):
    for j in range(0,cols):
        img[i][j][0] = Blending(x,img1[i][j][0],img[i][j][0])
        img[i][j][1] = Blending(x,img1[i][j][1],img[i][j][1])
        img[i][j][2] = Blending(x,img1[i][j][2],img[i][j][2])
cv2.imshow('Coverted Image',img)
cv2.imwrite("sol4-0.25.png", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
